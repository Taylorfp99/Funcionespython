#Ejercicio 7
Cree una función para contar la cantidad de números negativos de una lista.


````python
def neg(lst):
    asd=0
    for i in lst:
        if i<0:
            asd+=1
    return asd

nega=[1,2,3,4,-2,56,78,-27,-88,12,0]
neg(nega)        
````

# Ejercicio 8

Cree una función para contar la cantidad de números impares de una lista.


````python
def imp(lst):
    con=0
    for i in lst:
        if i%2!=0:
            con+=1
    return con

n=[1,2,3,4,2,56,78,27,88,12,0,1,7,77,33]     
imp(n)   
````
